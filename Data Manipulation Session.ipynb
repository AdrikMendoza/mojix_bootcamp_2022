```python
df.groupby(['continent', 'location']).count()[['new_deaths_smoothed']].rename(columns={'new_deaths_smoothed':'deaths_smoothed'})
```

```python
# Sorting

df_aux = df.groupby(['continent', 'location']).count()[['new_deaths_smoothed']].rename(columns={'new_deaths_smoothed':'deaths_smoothed'})
df_aux.sort_values('deaths_smoothed', ascending=False)
```

```python
df_aux_sorted = df_aux.sort_values('deaths_smoothed', ascending=False)
df_aux_sorted = df_aux_sorted.reset_index()
```

```python
# Pivot Table

df_aux_sorted['location'].nunique()
```

```python
df_aux_sorted['continent'].nunique()
```

```python
df_aux_sorted.head()
```

```python
# Pivot Table

df_aux_sorted.pivot_table(values='deaths_smoothed', index='location', columns='continent').fillna("")
```

```python
final = df_aux_sorted.pivot_table(values='deaths_smoothed', index='location', columns='continent').fillna("")
```

```python
df_a = final.reset_index()
df_a
```

```python
df_b = pd.DataFrame({'location':['Zimbabwe', 'Zambia', 'Albania'], 'red_flag':["yes", "yes", "TBD"]})
df_b
```

```python
df_merge = pd.merge(df_a, df_b, on='location', how='outer', indicator=True)
df_merge
```
